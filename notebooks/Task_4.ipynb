# Data Visualizations

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_train.csv")
```

## Bar Chart - Average Sale Price by Neighborhood

```python
neighborhood_avg = df.groupby('Neighborhood')['SalePrice'].mean().sort_values(ascending=False)

plt.figure(figsize = (14, 6))
neighborhood_avg.plot(kind = 'bar')
plt.title('Average Sale Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average SalePrice')
plt.xticks(rotation = 45)
plt.show()
```

Houses in Northridge are most expensive with average prices of 335,295 USD, while houses in Meadow Village are the cheapest at around 98,576 USD.

## Box Plot - SalePrice by OverallQual

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title('SalePrice Distribution by Overall Quality')
plt.show()
```

The box plot reveals a clear and consistent relationship between OverallQual and SalePrice. As quality ratings increase from 1 to 10, both the median sale price and the price range increase significantly. Quality-1 and quality-2 homes are tightly clustered below 100,000 USD, while quality-9 and quality-10 homes show median prices above 350,000 USD with considerably wider price ranges. This wider spread at higher quality levels suggests that while quality is a strong price driver, premium homes are also influenced by other factors such as location and size

## Scatterplot - GrLivArea vs SalePrice

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
plt.title('Living Area vs SalePrice')
plt.show()
```

The scatter plot shows a strong positive relationship between above-grade living area (GrLivArea) and SalePrice, consistent with the correlation of 0.71 identified in Task 3. Most homes fall between 500 and 2,500 square feet, selling between 50,000 USD and 300,000 USD. Notably, two large properties exceeding 4,500 square feet sold for unusually low prices, suggesting they may be distressed sales or properties with other significant drawbacks. A cluster of high-value outliers above $500,000 also suggests that factors beyond size, such as quality and location, contribute significantly to premium pricing.

```python
correlation = df.corr(numeric_only = True)['SalePrice'].sort_values(ascending = False)

top_positive = correlation.index[:5]
top_negative = correlation.index[-5:]
top_features = top_positive.append(top_negative)

plt.figure(figsize=(10, 5))
sns.heatmap(df[top_features].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Features Most Correlated with SalePrice')
```

## In Summary

The data visualizations in this task communicate four key insights about house pricing:
1. Neighborhood significantly impacts price. Houses in Northridge sell for an average of 335,295 USD, while houses in Meadow Village average just 98,576 USD.
2. Overall quality is the strongest visual predictor of price. Median SalePrice rises consistently from below 100,000 USD for quality-1 homes to above 350,000 USD for quality-10 homes.
3. Living area shows a strong positive relationship with price. Most homes between 500 and 2500 square feet sell between 50,000 USD and 300,000 USD, though anomalies suggest location and quality also play significant roles.
4. The correlation heatmap confirms OverallQual (0.79), GrLivArea (0.71), and GarageCars (0.64) as the three strongest predictors of SalePrice.
