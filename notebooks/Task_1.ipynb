### About the Dataset
The data shows descriptions of the houses that are put up for sale. One of the columns contains row ID (Id), seventy-nine of them are feature columns, and one of them is the target column(SalePrice).  Features cover a wide range of property characteristics including physical size (LotArea, GrLivArea), quality ratings (OverallQual, OverallCond), location (Neighborhood, MSZoning), extra features (MiscFeatures, MiscVal) and temporal information (YearBuilt, YrSold). It also contains descriptions of different elements of the houses such as exterior(Exterior 1st, ExterQual), fireplace(Fireplaces, FireplaceQu), roof(RoofStyle, RoofMat1), garage(GarageType, GarageQual)...etc
